Q1: Create a large random data file

```python
import random

rows, cols = 20000, 10000

with open("Q1_large_matrix.txt", "w") as f:
    for _ in range(rows):
        row = [str(random.randint(0, 100)) for _ in range(cols)]
        f.write(" ".join(row) + "\n")

print("Data file created successfully!")
```

Q2: WAP in python to Assign subsequent rows to Matrix first-row elements

```python
matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

result = {matrix[0][i]: [row[i] for row in matrix[1:]] for i in range(len(matrix[0]))}

print(result)
```

Q3: WAP to group Similar elements in a matrix

```python
matrix = [
    [1, 2, 2],
    [3, 1, 3],
    [2, 3, 1]
]

group = {}

for row in matrix:
    for val in row:
        group.setdefault(val, []).append(val)

print(group)
```

Q4: WAP to add and Subtract matrix in python

```python
A = [[1, 2], [3, 4]]
B = [[5, 6], [7, 8]]

rows, cols = len(A), len(A[0])

add = [[A[i][j] + B[i][j] for j in range(cols)] for i in range(rows)]
sub = [[A[i][j] - B[i][j] for j in range(cols)] for i in range(rows)]

print("Addition:", add)
print("Subtraction:", sub)
```

Q5:WAP To create an nxn sq matrix,where all sub-matrix has the sum of opposite corner elements as even

```python
import random

def create_matrix(n):
    matrix = [[random.randint(0, 9) for _ in range(n)] for _ in range(n)]

    for i in range(n - 1):
        for j in range(n - 1):
            if (matrix[i][j] + matrix[i+1][j+1]) % 2 != 0:
                matrix[i+1][j+1] += 1  # make sum even

    return matrix


n = 4
print(create_matrix(n))
```

Q6:WAP to Row-wise element addition in tuple matrix.

```python
A = ((1, 2, 3), (4, 5, 6))
B = ((7, 8, 9), (1, 2, 3))

result = tuple(
    tuple(A[i][j] + B[i][j] for j in range(len(A[0])))
    for i in range(len(A))
)

print(result)
```
